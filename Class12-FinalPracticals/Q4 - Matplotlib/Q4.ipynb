```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
q_values = []
# Values for p from 1 to 15
for i in range(0,16):
    q_values.append(i**2 - 2*i + 0.5)
    
# Plotting the line chart
plt.plot(range(0,16), q_values)

#(i) Give Chart Title: ‘Line chart for Equation'
plt.title('Give appropriate title for X and Y axis')
```

```python
#(ii) Give appropriate title for X and Y axis
plt.xlabel('domain p')
plt.ylabel('range q')
plt.plot(range(0,16), q_values)
```

```python
#(iii) Keep the width of line as 2
plt.plot(range(0,16), q_values, linewidth=2)
```

```python
#(iv) Display the line in red colour
plt.plot(range(0,16), q_values, color='red')
```
